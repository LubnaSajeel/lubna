```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
print(x)
print(y)
```

```python
#ONE HOT ENCODING
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
a=pd.DataFrame({'Actual value':y_test,'predicted_value':y_pred,'error':y_test-y_pred})
a
```

```python
from sklearn.metrics import mean_absolute_error
mean=mean_absolute_error(y_test,y_pred)
mean
```

```python
from sklearn.metrics import mean_absolute_percentage_error
per=mean_absolute_percentage_error(y_test,y_pred)
per
```

```python
from sklearn.metrics import mean_squared_error
squa=mean_squared_error(y_test,y_pred)
squa
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
c=np.sqrt(squa)
c
```
